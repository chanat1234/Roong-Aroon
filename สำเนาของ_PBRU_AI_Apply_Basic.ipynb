# พื้นฐานเกี่ยวกับชนิดตัวแปรแบบต่างๆ
image.png

```python
a = 5
b = 2.4
sum = a + b
print(sum)          # 7.4
print(type(sum))    # <class 'float'>
```

```python
# integer
num = 42
print(type(num), num)               # <class 'int'> 42

# float
float_num = 3.14159
print(type(float_num), float_num)   # <class 'float'> 3.14159

# boolean
bool_var = True
print(type(bool_var), bool_var)     # <class 'bool'> True

# string
string_var = "Hello, world!"
print(type(string_var), string_var) # <class 'str'> Hello, world!

# list
list_var = [1, 4, 2, 3]
print(type(list_var), list_var)     # <class 'list'> [1, 2, 3, 4]

# tuple
tuple_var = (5, 6, 7, 8)
print(type(tuple_var), tuple_var)   # <class 'tuple'> (5, 6, 7, 8)

# dictionary
dict_var = {'a': 1, 'b': 2, 'c': 3}
print(type(dict_var), dict_var)     # <class 'dict'> {'a': 1, 'b': 2, 'c': 3}
```

# การใช้งานตัวแปรขั้นสูง หรือตัวแปรแบบมีโครงสร้าง
image.png

```python
mylist = [1, 3, 2, 5, 4]    # ตัวแปร list ใช้สัญลักษณ์ [] เก็บค่า 1มิติ
print(mylist)               # [1, 3, 2, 5, 4]
print(mylist[:-1])          # [1, 3, 2, 5]
print(mylist[-1:])          # [4]
print(mylist[3])            # 5
```

```python
print(mylist[4])    # 4
mylist[4] = 9       # เปลี่ยนค่าใน index ที่ 4 จากเลข 4 เป็นเลข 9
print(mylist[4])    # 9
```

```python
for i in mylist:
    print(i)
```

```python
for i in range(0,5):
    print(mylist[i])
```

```python
mylist2 = [[1, 2, 3], [4, 5, 6], [7, 8, 9]] #ตัวแปร list ใช้สัญลักษณ์ [[],[],..] เก็บค่า 2มิติ
for i in mylist2:
    print(i)
```

```python
mylist2 = [[1, 2, 3], [4, 5, 6], [7, 8, 9]] #ตัวแปร list ใช้สัญลักษณ์ [[],[],..] เก็บค่า 2มิติ
for i in range(0,3):
    print(mylist2[i])
```

```python
import numpy as np

arr1 = np.array([1, 2, 3, 4, 5])            # สร้างอาเรย์ 1มิติ
arr2 = np.array([[7, 8, 9], [10, 11, 12]])  # สร้างอาเรย์ 2มิติ

print("One-dimensional array: ", arr1)
print("Two-dimensional array:\n", arr2)
print("Dimensions of arr1: ", arr1.ndim)
print("Dimensions of arr2: ", arr2.ndim)
print("Shape of arr1: ", arr1.shape)
print("Shape of arr2: ", arr2.shape)
print("Datatype of arr1: ", arr1.dtype)
print("Datatype of arr2: ", arr2.dtype)

# Reshape arr1 to have shape (1, 5)
arr1_reshaped1 = arr1.reshape((1, 5))
print("Reshape arr1 to have shape (1, 5)\n", arr1_reshaped1)

# Reshape arr1 to have shape (1, 5, 1)
arr1_reshaped2 = arr1.reshape((1, 5, 1))
print("Reshape arr1 to have shape (1, 5, 1)\n", arr1_reshaped2)

# Reshape arr1 to have shape (5, 1)
arr1_reshaped3 = arr1.reshape((5, 1))
print("Reshape arr1 to have shape (5, 1)\n", arr1_reshaped3)

# Reshape arr1 to have shape (5, 1, 1)
arr1_reshaped4 = arr1.reshape((5, 1, 1))
print("Reshape arr1 to have shape (5, 1, 1)\n", arr1_reshaped4)

# Reshape arr2 to have shape (1, 2, 3)
arr2_reshaped1 = arr2.reshape((1, 2, 3))
print("Reshape arr2 to have shape (1, 2, 3)\n", arr2_reshaped1)

# Reshape arr2 to have shape (3, 2)
arr2_reshaped2 = arr2.reshape((3, 2))
print("Reshape arr2 to have shape (3, 2)\n", arr2_reshaped2)

# Perform arithmetic operations on the arrays
arr1_reshaped = arr1.reshape(5, 1, 1) # Reshape arr1 to have shape (5, 1, 1) same shape as arr2

# Perform the addition operation between arr1_reshaped and arr2
result1 = arr1_reshaped + arr2
print(result1)

result2 = arr1_reshaped * arr2
print(result2)
```
